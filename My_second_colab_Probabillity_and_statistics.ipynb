```python
import numpy as np
import pandas as pd
import plotly.graph_objects as pg
import plotly.express as px
#Rolling od a Dice
print("The following is the probability of getting a number form 1 to 6 while rolling a dice")
def rolling_a_dice(roll=100):
  return pd.Series(np.random.choice(['1','2','3','4','5','6'],roll))
  # The above  line creates a pandas series object containig the results of the stimulation rolling of a dice
  #Ans all the six numbers are randomnly chosen using the funcion np.random.choice
results=rolling_a_dice()
print(results.value_counts()) # This function stores the individual count of each element in a dice  while rolling
res=pg.Figure([pg.Bar(x=results.value_counts().keys(),
                      y=results.value_counts().values)])
res.show()
#Tossing_a_coin
print("The following is the probability of getting either Heads or Tails while tossing a coin")
def Tossing_a_coin(toss=100):
  return pd.Series(np.random.choice(['Heads','Tails'],toss))
results=Tossing_a_coin()
print(results.value_counts())
res=pg.Figure([pg.Bar(x=results.value_counts().keys(),
                      y=results.value_counts().values)])
res.show()
```

```python
# Derived distribution is just like finding a new probability using the existing one
import numpy as np
import pandas as pd
import plotly.graph_objects as pg
import plotly.express as px
def Tossing_a_coin(toss=100):
  return pd.Series(np.random.choice(['Heads','Tails'],toss))
results=pd.Series([np.sum(Tossing_a_coin(500)=='Heads') for _ in range(10000)])
#This piece of code performs tossing of a coin for 20 times and adds the events where the coin shows up tails
# this whole process is repeated 100 times
res=pg.Figure([pg.Bar(x=results.value_counts().keys(),
                      y=results.value_counts().values)])
res.show()
```

```python
#Refgression to mean example for throwing a dice
import numpy as np
import pandas as pd
import plotly.graph_objects as pg
import plotly.express as px
print('Regression to mean for throwing a  dice :')
def throw_dice(number_of_tosses=100):
    return pd.Series(np.random.randint(1, 6+1, number_of_tosses))
results = {shots: throw_dice(shots) for shots in [10, 100, 1000, 10000, 1000000]}
fig = pg.Figure([pg.Bar(x=result.value_counts().keys(),
                        y=result.value_counts().values / shots,)
                    for shots, result in results.items()])
fig.add_shape(type='line', x0=0.5, y0=1/6, x1=6.5, y1=1/6, line=dict(color='indigo',))
fig.show()
#Regression to mean for tossing a coin
print('Regression to mean for Tossing a  coin  :')
def Tossing_a_coin(toss=100):
  return pd.Series(np.random.choice(['Heads','Tails'],toss))
results={tosses: Tossing_a_coin(tosses) for tosses in [10,100,1000,10000]}
fig = pg.Figure([pg.Bar(x=result.value_counts().keys(),
                        y=result.value_counts().values / tosses,)
                    for tosses, result in results.items()])
fig.add_shape(type='line', x0=0.5, y0=1/2 , x1=6.5 ,y1=1/2,line=dict(color='Black',))
fig.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as pg
import plotly.express as px
def numbered_cards_before_ace(trails=50 ,show_trails=5):
  values=['A','2','3','4','5','6','7','8','9','J','Q','K']
  ranks=['H','D','S','C']
  thrown_cards_count=[]
  for exp in range(trails):
     deck_of_cards = [value + rank for value in values for rank in ranks]
     cards_thrown=0
     cards_pile=[]
     while len(deck_of_cards) > 0:
      picked_card=np.random.choice(deck_of_cards)
      cards_pile.append(picked_card)
      if picked_card[0]=='A':
        break
      cards_thrown = cards_thrown+1
      deck_of_cards.remove(picked_card)
      number=1
      if exp < show_trails:
        print(cards_thrown,'cards were thrown :',' '.join(cards_pile))
      thrown_cards_count.append(cards_thrown)
  return np.mean(thrown_cards_count),thrown_cards_count
expected_cards_thrown,cards_thrown_count=numbered_cards_before_ace(1000)
number=0
print(f'\n{expected_cards_thrown} cards are tossed before we see the first ace!')
results = pd.Series(cards_thrown_count)
fig = pg.Figure([pg.Bar(x=results.value_counts().keys(),
                        y=results.value_counts().values)])
fig.show()
```

```python
#Prbobailty Distributions:
# 1)Binomial Distribution
# It is a process of getting k wins in n trails
from scipy.stats import binom
binom.pmf(k=5, n=5, p=0.25)#pmf:probability mass function
# The above function calculates  the binomial distribution function for a game
# , for which the probability of loosing is 0.25 and we are trying 5 matches
# since all matches are independent the pmf is 0.25*0.25*0.25*0.25*0.25
# Probability of winning 3 matches in 5
binom.pmf(k=3,n=5,p=0.75)
```

```python
# 2) Normal Distribution function
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100) #creates an array with evenly spaced values (100) between -10 and 10
f1 = norm.pdf(x, loc=0, scale=1)# This calulates the probability density function for mean=0(loc) and standard deviation =1(scale)
f2 = norm.pdf(x, loc=1, scale=1)
f3 = norm.pdf(x, loc=0, scale=2)

plt.plot(x, f1, label="$\mu=0, \sigma=1$")#plotting of values of x on the graph
plt.plot(x, f2, label="$\mu=1, \sigma=1$")
plt.plot(x, f3, label="$\mu=0, \sigma=2$")
plt.vlines(x=0, ymin=0, ymax=0.5,linestyles='dotted')

plt.legend()
plt.title("$\mu$ and $\sigma$ affect the shape of the normal distribution")
plt.show()
```

```python
#Poisson Distribution function
from scipy.stats import poisson
x = np.arange(10)
f1 = poisson.pmf(x, mu=1)
f2 = poisson.pmf(x, mu=2)
f3 = poisson.pmf(x, mu=3)

plt.bar(x, f1, label="$\lambda=1$")
plt.legend()
```

```python
x @ f1
plt.bar(x, f2, label="$\lambda=2$")
plt.legend()
x @ f2
plt.bar(x, f3, label="$\lambda=3$")
plt.legend()
x @ f3
```
